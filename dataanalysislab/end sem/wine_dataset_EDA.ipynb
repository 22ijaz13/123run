```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("winequality-red.csv")
print(data)
```

```python
cols=list(data.columns)
print(cols)
```

```python
#BOX PLOT
data.boxplot(column=["residual sugar"])
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
plt.title("Box plot for residual sugar")
plt.show()
```

```python
for i in cols:
    data.boxplot(column=[i])
    plt.show()
```

```python
import seaborn as sns

sns.boxplot(data=data["residual sugar"])
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
plt.title("Box plot for total sulphur dioxide")
plt.show()
```

```python
column_data=data['residual sugar']
q3=column_data.quantile(0.75)
q1=column_data.quantile(0.25)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliers=column_data[(column_data<lower_bound)|(column_data>upper_bound)]
print("Number of outliers:", len(outliers))

data.boxplot(column=['residual sugar'])
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
plt.title("Box plot for residual sugar")
plt.show()
```

```python
outliers_df=data.loc[outliers.index]
print(outliers_df)
```

```python
data_filtered=data.drop(outliers.index)
print(data_filtered)
```

```python
column_data=data['alcohol']
q1=column_data.quantile(0.25)
q3=column_data.quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outlierss=column_data[(column_data>upper_bound)|(column_data<lower_bound)]
pdata=data[~data['alcohol'].isin(outlierss)]
adata=data.drop(outliers.index)
#print("Count after:",len(pdata))
print("Count after:",len(adata))
print("Count before:",len(data))
```

```python
numeric_data=data.select_dtypes(include=['number'])
covariance_matrix=data.cov()
correlation_matrix=data.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

```python
print(data['alcohol'].skew())
print(data['alcohol'].kurtosis())
```

```python
alcohol_data=data['alcohol']
plt.figure(figsize=(6, 4))
plt.hist(alcohol_data, bins=30, color='skyblue', alpha=0.7)
plt.title('Skewness = {:.2f}'.format(alcohol_data.skew()))
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()

# Plotting kurtosis
plt.figure(figsize=(6, 4))
plt.hist(alcohol_data, bins=30, color='salmon', alpha=0.7)
plt.title('Kurtosis = {:.2f}'.format(alcohol_data.kurtosis()))
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```
